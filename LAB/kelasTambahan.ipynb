```python
# Riemann Integral
import numpy as np

# siapkan function 
def f(x):
    return 4*(x**4) + 2*(x**3) - 6*x

# siapkan lower limit and upper limit 
x1 = -3 
x2 = 3

# siapkan variable jumlah titik 
# jumlah persegi = jumlha titik -1 
n = 30 

# cari lebar tiap persegi 
lebar = (x2-x1)/(n-1)

# Menentukan koordinat x & y
x = np.linspace(x1, x2, n)
y = f(x)

# Left Riemann 
leftRiemann = lebar * np.sum(y[0:n - 1])
print(f"Left Riemann : {leftRiemann}")

# Right Riemann 
rightRiemann = lebar * np.sum(y[1:n])
print(f"Right Riemann : {rightRiemann}")

# Mid Reimann 
x_mid = (x[0:n-1]+x[1:n])/2
y_mid = f(x_mid)

midRiemann = lebar * np.sum(y_mid)
print(f"Middle Riemann : {midRiemann}")
```

```python
# Trapezoid 
import numpy as np 

a = -np.pi
b = 2*np.pi
n = 27


def f (x): 
    return ((np.sin(x))**2 + (np.sin(x)) +2)

lebar = (b - a) / (n - 1)
x = np.linspace(a, b, n)
y = f(x)

trapezoid = lebar / 2 * (y[0] + 2 * np.sum(y[1:n-1]) + y[n-1])
print(f"Trapezoid : {trapezoid}")

```

```python
# Newton raphson
import numpy as np 

def f(x): 
    return x**6 +2*(x**2)-3

def df(x): 
    return 6*(x**5) + 4*(x)

# buat funciton newthon raphson 
def newtonRaphson(x, eps = 0.01, maxIter = 15):
    for i in range(maxIter):
        x = x - f(x) / df(x)
        print(f"Iteration of {i+1}. new root = {x}")
        if (np.abs(f(x) < eps)):
            print(f"The root of the equation is {x}")
            return
    print("Error, max iterations reached")

newtonRaphson(4)
        
    
```
